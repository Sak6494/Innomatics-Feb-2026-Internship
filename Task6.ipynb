```python
#2)Online Food Delivery Time Estimator
def delivary(Distance, weather,Traffic):
    Speed=40 # km per hour
    Base_Time =Distance/Speed
    if(Traffic=="MEDIUM"):#Let Medium Traffic delay adds 60%
        Traffic_Delay=Base_Time*(0.6)
        Time_After_Traffic =Base_Time+Traffic_Delay
    elif(Traffic=="HIGH"):#Let High Traffic delay adds 120%
        Traffic_Delay=Base_Time*(1.2)
        Time_After_Traffic =Base_Time+Traffic_Delay
    elif(Traffic=="LOW"):
        Time_After_Traffic=Base_Time
    
    if(weather=="RAINY"):
         Delay=Time_After_Traffic*(0.5 )#50%
         Final_time=Time_After_Traffic+Delay
    elif(weather=="HEAVY STORM"):
         Delay=Time_After_Traffic*(1.2) #120%
         Final_time=Time_After_Traffic+Delay
    elif(weather=="SUMMER"):
         Delay=Time_After_Traffic
         Final_time=Time_After_Traffic+Delay
    final_time_min=Final_time*60
    fixed_overhead =15#for food pickup
    Estimated_Time = fixed_overhead+ final_time_min
    print(f"Estimated Delivery Time:{round(Estimated_Time )} minutes")
Distance=int(input("Distance:"))
Traffic =input("Traffic Leve:").upper()
weather=input("Weather:").upper()
    
delivary(Distance, weather,Traffic)  
        
        
```

```python
#3)Movie theatre seat Occupancy

def Booking(Total_seats,Booked_Seats):


    booked_count=len(Booked_Seats)
    
    occupancy=(booked_count/Total_seats)*100
    if(occupancy>=75):
        Status ="Almost Full"
    elif(occupancy==100):
        Status = "Housefull"
    else:
        Status ="Seats Available"
   
    if(occupancy>=80):
        shows ="Consider opening additional shows"
    else:
        shows="No additional shows needed"
    print(f"Occupancy:{occupancy}%")
    print(f"Show Status:{Status}")
    print(f"Shows:{shows}")
    
Total_seats=200
Booked_Seats=[1] * 150


Booking(Total_seats,Booked_Seats)
```

```python
#4)Cloud server load classification System
def classify_server_load(cpu_readings):
    total = 0

    
    for reading in cpu_readings:
        total += reading

    # Average CPU usage
    average_cpu = total / len(cpu_readings)

  
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu < 80:
        status = "Warning"
    else:
        status = "Critical"

    return {
        "Average CPU Load": f"{average_cpu:.0f}%",
        "Server Status": status
    }
cpu_readings = [45, 60, 70, 85, 90]

result = classify_server_load(cpu_readings)
print(result)

    
```
