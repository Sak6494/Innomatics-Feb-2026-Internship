```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]
    
    # Check minimum recharge amount
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    
    # Check if selected plan is valid
    if plan not in valid_plans:
        return "Invalid plan selected."
    
    return "Recharge Successful!"

# Retry system using WHILE loop
while True:
    amount = int(input("Enter recharge amount: "))
    plan = int(input("Enter plan (199/299/399/599): "))
    
    result = validate_recharge(amount, plan)
    print(result)
    
    if result == "Recharge Successful!":
        break
    else:
        print("Please try again.\n")
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

inventory = {
    "Laptops": 10,
    "Keyboards": 25,
    "Mouse": 12,
    "Monitors": 18
}

check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System
def res(marks):
    total = 0
    
    # Calculate total marks
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks_list = [60, 45, 70, 55]
result = res(marks_list)

print("Result:", result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    
    # Add 25% extra during peak hours
    if peak.lower() == "yes":
        fare += fare * 0.25
    
    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")
    
    total_fare = fare(distance, peak)
    print(f"Total Fare: ₹{total_fare:.2f}")
    
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def check(att_list):
    pr_count = 0
    
    # Count present days
    for day in att_list:
        if day == "P":
            pr_count += 1
    
    prcnt = (pr_count / len(att_list)) * 100
    
    if prcnt >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P"]
status = check(attendance)

print("Attendance Status:", status)
```
