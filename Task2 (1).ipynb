```python
#2)Store employee names and their salaries, and determine which employee has the highest salary
employee ={
    "Ravi":75000,
    "Anita":68000,
    "Kiran":72000
}
sal=0
name=""
for emp,salary in employee.items():
    if(sal < salary):
        sal=salary
        name=emp
    
print(name)
```

```python
#3)Find the maximum and minimum values from a list of numbers.
numbers = [45, 22, 89, 10, 66]
max_num =0

for i in numbers:
    if(i>max_num):
        max_num=i
min_num=max_num
for i in numbers:
    if(min_num>i):
        min_num=i

print(f"maximum no is:{max_num} and minimum no is:{min_num}")
        
```

```python
#4)Count how many products have a price greater than 1000.
prices = [450, 1200, 899, 1500, 300]
count =0
for price in prices:
    if(price>1000):
        count+=1
print(f"count of products having price greater than 1000 is {count}")
```

```python
#5)Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).
attendance = ["P", "P", "A", "P", "P"]
count =0
for i in attendance:
    if(i=="P"):
        count+=1
per_att=(count/len(attendance))*100
print(f" Attendance Percentage {per_att}")
        
```

```python

#6)You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.

phone_numbers = [9876543210, 9123456789, 9876543210]
set1=set()
for i in phone_numbers:
    set1.add(i)
print(f"Unique phone numbers are:{set1}")
```

```python
#7)Count Character Frequency

string = input("Enter a string:")

dict ={}

for i in string:
    if(i in dict):
        dict[i]+=1
    else:
        dict[i]=1
       
print(dict)
```
