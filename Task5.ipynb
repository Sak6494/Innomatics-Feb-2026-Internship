```python
#1)Social Media-Post Engagement Analyzer

def social_media(like_list):
    total_like = 0
    for like in like_list:
        total_like+=like
    if(total_like>=1000):
        status= ("Viral Post")
    else:
        status = ("Normal Engagement")
    print(f"Total Likes:{total_like}")
    print(f"Post Status:{status}")

like_list=[30,400,200]

social_media(like_list)    

        
        
    
    
```

```python
#3)Agriculture-Rainfall Adequacy Cheker
def Rain(Data_list,Required):
    sum1=0
    avg=0
    
    for data in Data_list:
        sum1+=data
    avg=sum1/len(Data_list)
    if(avg  >=Required):
        status =("Adquate Rainfall")
    else:
        status=("Inaquate Rainfall")
    print(f"Average Rainfall:{avg}")
    print(f"Rainfall Status:{status}")
Required= int(input("What is required level of Rainfall:"))
Data_list=[5,7,10,15]
Rain(Data_list,Required)
```

```python
#4)Social Media -Duplicate Account Detecton
list1=["Ravi","Om","shiva","Ravi"]
Set1 =set()

def Check_dup(list1,Set1,count):
    for data in list1:
        if(data in Set1):
            count+=1
        elif(data not in Set1):
            Set1.add(data)
    if(count>0):
        status = "Yes"
    else:
        status="No"
    print(f"Duplicate Accounts Found:{status}")
        
count=0
Check_dup(list1,Set1,count)

    
```

```python
#7)System Monitoring- Application Health

def application(App_Health,count):
    for i in App_Health:
        if(i=="error"):
            count+=1
    if(count==0):
        status = "Healthy"
    elif(count<=5):
        status = "Minor Issues"
    else:
        status ="Critical Issue"
    print(f"Error Count:{count}")
    print(f"System Status: {status}")
App_Health = ["error","working","working"]
count=0
application(App_Health,count)  
```

```python
#8)Banking-Daily Transcation Limit Checker

def Banking(Daily_limit,amount):
    if(amount<=Daily_limit):
        status = "Approved"
    else:
        status = "Rejected"
    print(f"Transaction Amount:{amount}")
    print(f"Transaction Status:{status}")
amount=input("Enter the amount:")
Daily_limit=50000
Banking(Daily_limit,amount)
    
```
