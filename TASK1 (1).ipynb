```python
#2)Analyze student results.
marks=[45,78,90,33,60]
Pass=0
Fail=0
for i in marks:
    if(i>=50):
        Pass+=1
    else:
        Fail+=1
print(f"{Pass} Students Are Pass and {Fail} Students are Fail In Exam")
```
