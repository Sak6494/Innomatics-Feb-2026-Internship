```python

#1)Smart Light Controller

class Display:
    status ="OFF"
    def __init__(self,Action):
        if(Action =="ON" ):
            self.status ="ON"
        else:
            self.status =status
Light_Name=input("Light Name:")
action = input("Action:")
l1 = Display(action)
print(f"{Light_Name} is {l1.status}")

```

```python
#2) Employee ID Card System
class Student:
    def __init__(self,Name,ID,Department):
        self.Name=Name
        self.ID=ID
        self.Department=Department
    def Display(self):
        print("Employee ID Card")
        print(f"Name:{self.Name}")
        print(f"ID:{self.ID}")
        print(f"Departmnet:{self.Department}")
    
Employee_Name=input("Employee Name:")
Employee_ID=input("Employee ID:")
Department =input("Department:")
st1=Student(Employee_Name,Employee_ID,Department)
st1.Display()
        
```

```python
#3)Mobile Contact Record
class Contact:
    def __init__(self,Name,Number):
        self.Name=Name
        self.Number=Number
    def Display(self):
        print("Contact Saved")
        print(f"Contact Name:{self.Name}")
        print(f"Phone Number:{self.Number}")
Contact_Name = input("Contact Name:")
Phone_Number =input("Phone Number:")
Mb=Contact(Contact_Name,Phone_Number)
Mb.Display()
```

```python
#4)Product Price  Tag Grenerator

class Product:
    def __init__(self,Name,Price):
        self.Name=Name
        self.Price=Price
    def Display(self):
        
        print(f"Product:{self.Name}")
        print(f"Price:{self.Price}")
Product_Name = input("Product Name:")
cost =input("Price:")
pro1=Product(Product_Name,cost)
pro1.Display()
```

```python
#5)MovieRating Display System
class Movie:
        def __init__(self,Name,Rating):
            self.Name=Name
            self.Rating=Rating
        def Display(self):
        
            print(f"Movie:{self.Name}")
            print(f"Rating:{self.Rating}/5")
Movie_Name = input("Movie Name:")
Ratings =input("Rating:")
Movie1=Movie(Movie_Name,Ratings)
Movie1.Display()
    
```

```python
#6)Delivery Address Manager
class Delivery:
        def __init__(self,Name,Address):
            self.Name=Name
            self.Address=Address
        def Display(self):
            print("Delivery Details")
            print(f"Customer :{self.Name}")
            print(f"Address:{self.Address}")
Customer_Name = input("Customer Name:")
Add =input("Address:")
Delivery1=Delivery(Customer_Name,Add)
Delivery1.Display()
```
