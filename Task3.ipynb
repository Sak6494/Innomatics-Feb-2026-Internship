```python
#1)Employee Performance Bonus Eligibility


employees={
    "Ravi":92,
    "Anita":88,
    "Kiran":92,
    "Suresh":85,
    
}
higest_scor = 0
Emp_name=[]
for name, performence in employees.items():
    if(higest_scor <= performence):
        higest_scor = performence
        Emp_name.append(name)
    
print(f"Top performers Eligible for bonus:{Emp_name} (Score:{higest_scor})")
    
```

```python
#2) Search Query Keyword Analysis
import string
query =input('Query of  customer:')
query  = query.lower().translate(str.maketrans('','',string.punctuation))
Processed_query=query.split()
summary={}
for i in Processed_query:
     if(i in summary):
         summary[i]+=1
     else:
         summary[i]=1
result={}
for que,val in summary.items():
    if(val>1):
       result[que]=val
print(result)

```

```python
#3): Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
new_data=[]
for reading in sensor_readings:
    if(reading%2==0):
        new_data.append((sensor_readings.index(reading),reading))
        
        
print(f"Valid Sensor Readings (Hour,Value): {new_data}")     
```

```python
#4) Email Domain Usage Analysis


emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

emails_analysis=[]
for email in emails:
    email =email.split("@")
    emails_analysis.append(email)
    gmails_count =0
    yahoo_count=0
    for i in emails_analysis:
        if(i[1]=="gmail.com"):
            gmails_count+=1
        else:
            yahoo_count+=1
print(f"gmail.com: {(gmails_count/len(emails_analysis))*100} yahoo.com: {yahoo_count/len(emails_analysis)*100}")
```

```python
#5)Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sales=0
for i in sales:
    avg_sales = sum(sales)/len(sales)
    if(i>avg_sales):
        print(f"Day {sales.index(i)}:{i}")
    
```

```python
#6) Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
user={}
for i in user_ids:
    if(i in user):
        user[i]+=1
    else:
        user[i]=1
for users,vall in user.items():
    if(vall>1):
        print(f"{users}->{vall} times")
    
    
```
